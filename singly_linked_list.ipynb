```python
class Node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next
```

```python
class SLinkedList:
    def __init__(self):
        self.start = None
    
    def insert_at_beginning(self, data):
        node = Node(data, self.start)
        self.start = node

    def insert_at_middle(self,data,pos):
        if pos<=0 or pos>self.get_length():
            print('invalid entry...')
        elif pos==1:    self.insert_at_beginning(data)
        else:
            p=self.start
            q=None
            count=0
            node = Node(data)
            while p!=None:
                count+=1
                if count==pos:
                    node.next=p
                    q.next=node
                    break
                q=p
                p=p.next

    def insert_at_end(self, data):
        node = Node(data)
        if self.start == None:
            self.start = node
        else:
            p=self.start
            while p.next!=None:
                p=p.next
            p.next = node

    def get_length(self):
        c=0
        p=self.start
        while p!=None:
            c+=1
            p=p.next
        return c

    def delete_at_beginning(self):
        if self.start == None:
            print('list is empty')
        else:
            self.start = self.start.next

    def delete_at_middle(self, pos):
        if pos<=0 or pos>self.get_length():
            print('invalid entry...')
        elif pos==1:    self.delete_at_beginning()
        else:
            p=self.start
            q=None
            c=0
            while p!=None:
                c+=1
                if c==pos:
                    q.next = p.next
                    break
                q=p
                p=p.next

    def delete_at_end(self):
        if self.start==None:
            print('list is empty')
        elif self.start.next==None:
            self.start=None
        else:
            p=self.start
            q=None
            while p.next!=None:
                q=p
                p=p.next
            q.next=None

    def traverse(self):
        if self.start == None:  print('list is empty')
        else:
            p=self.start
            while p!=None:
                print(p.data, end='->')
                p=p.next
```
